```python
import stanza
import spacy_stanza
import pandas as pd
from tqdm.auto import tqdm
```

```python
stanza.download('ru')
stanza.download('sv')
```

```python
nlp_ru = stanza.Pipeline('ru', processors='tokenize')
nlp_sw = stanza.Pipeline('sv', processors='tokenize')
```

```python
# Uploading the dataset
df = pd.read_csv('change_translated.csv')
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
# Tokenizing the examples
russian = []
swedish = []
for sent_ru, sent_sw in zip(df.sentence.to_list(), df.swedish_sentence.to_list()):
    doc_ru = nlp_ru(sent_ru)
    doc_sw = nlp_ru(sent_sw)
    r = []
    s = []
    for sent in doc_ru.sentences:
        r.extend([token.text for token in sent.tokens])
    russian.append(' '.join(r))
    for sent in doc_sw.sentences:
        s.extend([token.text for token in sent.tokens])
    swedish.append(' '.join(s))
```

```python
# Writing the file for the alignment
with open('new_pairs.txt', 'w') as fw:
    for ru, sw in zip(russian, swedish):
        fw.write(ru.lower() + ' ||| ' + sw.lower())
        fw.write('\n')
```

## Verb extraction

To analyse the field of change in Swedish we should know which verbs correspond to the Russian pre-defined ones. 

```python
nlp_sw = spacy_stanza.load_pipeline("sv")
```

```python
with open('new_pairs_clean.txt', encoding='utf-8') as fh:
    paired_sentences = fh.readlines()
paired_sentences = [(i.strip().split(' ||| ')[0].split(), i.strip().split(' ||| ')[1].split()) \
                      for i in paired_sentences]
```

```python
# Create the dictionary with Russian token index as a key and Swedish token indices as a value

correspondence = []
with open('aligned_sentences.txt') as fh:
    aligned_sent = fh.readlines()
for line in aligned_sent:
    line_raw = line.strip().split()
    cor_dic = {}
    for i in line_raw:
        if int(i.split('-')[0]) not in cor_dic.keys():
           cor_dic[int(i.split('-')[0])] = []
        cor_dic[int(i.split('-')[0])].append(int(i.split('-')[1]))
    correspondence.append(cor_dic)
```

```python
all_verbs = []
for cor, pair, verb in tqdm(zip(correspondence, paired_sentences, df['form_verb'].to_list()), 
                            total=len(paired_sentences)):
    ru, sw = pair
    try:
        idx = cor[ru.index(verb.strip())]
        v = '–'
        for i in idx:
            if nlp_sw(sw[i])[0].pos_ == "VERB":
                v = nlp_sw(sw[i])[0].lemma_  # Collect lemma if the token is a verb (it is useful for phrasal verbs e.g. byta ut)
        all_verbs.append(v)
    except Exception as e:  # To check all the cases where the verb could not be extracted
        print()
        all_verbs.append('–')
        print(verb)
        print(e)
        print(ru)
        print(sw)
        print(cor)
```

```python
df['lemma_verb_sw'] = all_verbs
df.to_csv('change_verbs.csv')
```
